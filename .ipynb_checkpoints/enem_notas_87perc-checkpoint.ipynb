### Importando dados de treino

```python
import pandas as pd
pd.set_option('display.max_columns', None)

df_train = pd.read_csv('train.csv')

df_train.head()
```

### Criando novo dataframe, somente com as colunas que serão usadas

```python
df_x = df_train[['CO_UF_RESIDENCIA',
                'NU_IDADE',
                'TP_SEXO',
                'TP_COR_RACA',
                'TP_NACIONALIDADE',
                'TP_ESCOLA',
                'TP_ENSINO',    
                'IN_TREINEIRO',
                'TP_DEPENDENCIA_ADM_ESC',
                'IN_BAIXA_VISAO',
                'IN_CEGUEIRA',
                'IN_SURDEZ',     
                'IN_DISLEXIA',    
                'IN_DISCALCULIA',
                'IN_SABATISTA',  
                'IN_GESTANTE',   
                'IN_IDOSO',
                'Q001',   
                'Q002',  
                'Q006',  
                'Q024',  
                'Q025',  
                'Q026',  
                'Q027',  
                'Q047',  
                'TP_LINGUA',
                'TP_PRESENCA_CN',
                'TP_PRESENCA_CH',
                'TP_PRESENCA_LC',
                'CO_PROVA_CN',   
                'CO_PROVA_CH',  
                'CO_PROVA_LC',  
                'CO_PROVA_MT',  
                'NU_NOTA_CN',  
                'NU_NOTA_CH',   
                'NU_NOTA_LC',
                'NU_NOTA_MT',
                'TP_STATUS_REDACAO',
                'NU_NOTA_COMP1', 
                'NU_NOTA_COMP2',
                'NU_NOTA_COMP3',
                'NU_NOTA_COMP4',
                'NU_NOTA_COMP5',
                'NU_NOTA_REDACAO',
                'TP_ST_CONCLUSAO']]

df_x.head()
```

### Verificando valores missing

```python
from quilt.data.ResidentMario import missingno_data
import missingno as msno

mt = msno.matrix(df_x)
```

### Substituindo valores NaN por zero

```python
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'

#dftr['TP_SEXO'] = dftr['TP_SEXO'].fillna(0);
#dftr['TP_ENSINO'] = dftr['TP_ENSINO'].replace(np.nan, 0)

df_x = df_x.replace(np.nan, 0)

mtx = msno.matrix(df_x)

df_x.head(10)
```

### Tratamento para conversão dos valores hexa para int

```python
coProvaCh, uniques = pd.factorize(df_x.CO_PROVA_CH)
df_x['CO_PROVA_CH'] = coProvaCh
#print(coProvaCh)

tpSexo, uniques = pd.factorize(df_x.TP_SEXO, sort=True)
df_x['TP_SEXO'] = tpSexo

#cols = [df_x['Q001'], df_x['Q002'], df_x['Q006'], df_x['Q024'], df_x['Q025'], df_x['Q026'], df_x['Q027'], df_x['Q047']]
#fact = map(lambda x: pd.factorize(x, sort=True), cols)

df_x['Q001'], uniques = pd.factorize(df_x.Q001, sort=True)
df_x['Q002'], uniques = pd.factorize(df_x.Q002, sort=True)
df_x['Q006'], uniques = pd.factorize(df_x.Q006, sort=True)
df_x['Q024'], uniques = pd.factorize(df_x.Q024, sort=True)
df_x['Q025'], uniques = pd.factorize(df_x.Q025, sort=True)
df_x['Q026'], uniques = pd.factorize(df_x.Q026, sort=True)
df_x['Q027'], uniques = pd.factorize(df_x.Q027, sort=True)
df_x['Q047'], uniques = pd.factorize(df_x.Q047, sort=True)

df_x['CO_PROVA_CN'], uniques = pd.factorize(df_x.CO_PROVA_CN, sort=True)
df_x['CO_PROVA_LC'], uniques = pd.factorize(df_x.CO_PROVA_LC, sort=True)
#df_x['CO_PROVA_MT'], uniques = pd.factorize(df_x.CO_PROVA_MT, sort=True)

del df_x['CO_PROVA_MT']

df_x.head(10)
```

### Criação do modelo de regressão linear

```python
from sklearn import linear_model

x = df_x
y = df_x['NU_NOTA_MT']

lm = linear_model.LinearRegression()
model = lm.fit(x, y)

model.score(x, y)
```

### Carregando o dataset de test

```python
df_test = pd.read_csv('test.csv')

df_test.head()
```

### Substituindo valores NaN por zero

```python
df_test = df_test.replace(np.nan, 0)

mtx = msno.matrix(df_test)

df_test.head(10)
```

### Tratamento para conversão dos valores hexa para int

```python
df_test['TP_SEXO'], uniques = pd.factorize(df_test.TP_SEXO, sort=True)
df_test['Q001'], uniques = pd.factorize(df_test.Q001, sort=True)
df_test['Q002'], uniques = pd.factorize(df_test.Q002, sort=True)
df_test['Q006'], uniques = pd.factorize(df_test.Q006, sort=True)
df_test['Q024'], uniques = pd.factorize(df_test.Q024, sort=True)
df_test['Q025'], uniques = pd.factorize(df_test.Q025, sort=True)
df_test['Q026'], uniques = pd.factorize(df_test.Q026, sort=True)
df_test['Q027'], uniques = pd.factorize(df_test.Q027, sort=True)
df_test['Q047'], uniques = pd.factorize(df_test.Q047, sort=True)
df_test['CO_PROVA_CH'], uniques = pd.factorize(df_test.CO_PROVA_CH, sort=True)
df_test['CO_PROVA_CN'], uniques = pd.factorize(df_test.CO_PROVA_CN, sort=True)
df_test['CO_PROVA_LC'], uniques = pd.factorize(df_test.CO_PROVA_LC, sort=True)
#df_test['CO_PROVA_MT'], uniques = pd.factorize(df_test.CO_PROVA_MT, sort=True)

# Guardando o NU_INSCRICAO 
answer = df_test['NU_INSCRICAO']

# Movendo o dataset_test para outra variavel
del df_test['NU_INSCRICAO']
del df_test['SG_UF_RESIDENCIA']
del df_test['CO_PROVA_MT']

answer.head(10)
```

```python
df_test.head(10)
```

### Predição em cima do modelo

```python
pred = model.predict(df_test)

df_csv = pd.DataFrame(answer)
pred_csv = pd.DataFrame(pred)

df_csv['NU_NOTA_MT'] = round(pred_csv, 2)

for i, row in df_csv.iterrows():
    if row['NU_NOTA_MT'] < 50:
        df_csv.at[i,'NU_NOTA_MT'] = 0

df_csv.head(10)
```

### Exportando resultado para CSV

```python
df_csv.to_csv('answer_87.csv', index = False, header=True)
```
